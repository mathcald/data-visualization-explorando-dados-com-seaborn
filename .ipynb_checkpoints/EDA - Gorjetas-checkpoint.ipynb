# Importando base de dados

```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('dados/tips.csv')
df_traduzido = pd.read_csv('dados/gorjetas.csv')
```

```python
df.head()
```

```python
df.info()
```

# Traduzindo as colunas e campos

```python
# Conferindo as colunas
df.columns
```

```python
renomear = {
    'total_bill': 'valor_da_conta',
    'tip': 'gorjeta', 
    'dessert': 'sobremesa', 
    'day': 'dia_da_semana', 
    'time': 'hora_do_dia', 
    'size': 'tota_de_pessoas'
}
```

```python
df_gorjetas = df.rename(columns = renomear)
```

```python
df_gorjetas.head()
```

```python
df_gorjetas['dia_da_semana'].unique()
```

```python
dias = {
    'Sun': 'Domingo', 
    'Sat': 'SÃ¡bado', 
    'Thur': 'Quinta', 
    'Fri': 'Sexta'
}
```

```python
df_gorjetas['dia_da_semana'] = df_gorjetas['dia_da_semana'].map(dias)
```

```python
df_gorjetas.head()
```

```python
df_gorjetas['sobremesa'].unique()
```

```python
sim_nao = {
    'No': 'NÃ£o', 
    'Yes': 'Sim'
}
```

```python
df_gorjetas['sobremesa'] = df_gorjetas['sobremesa'].map(sim_nao)
```

```python
df_gorjetas.head()
```

```python
df_gorjetas['hora_do_dia'].unique()
```

```python
hora = {
    'Dinner': 'Jantar',
    'Lunch': 'AlmoÃ§o'
}
```

```python
df_gorjetas['hora_do_dia'] = df_gorjetas['hora_do_dia'].map(hora)
```

```python
df_gorjetas.head()
```

```python
df_gorjetas.to_csv('dados/gorjetas.csv', index=False)
```

## Teste de hipÃ³tese

```python
from scipy.stats import ranksums
```

```python
df_traduzido.query("sobremesa == 'Sim'").porcentagem
```

```python
sobremesa = df_traduzido.query("sobremesa == 'Sim'").porcentagem
sem_sobremesa = df_traduzido.query("sobremesa == 'NÃ£o'").porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
print('O valor do p-value Ã© {}'.format(r.pvalue))
```
